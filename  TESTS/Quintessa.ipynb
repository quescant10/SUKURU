```python

# first neural network with keras make predictions
from numpy import loadtxt
import logging
from colorama import *
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# load the dataset
print("Quintessa LOADING...")
dataset = loadtxt('/Users/aiydenkaidon/shangi/CSV/pima-indians-diabetes.csv', delimiter=',')
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
# define the keras model
logging.basicConfig(filename='QUINTESSAlog', encoding='utf-8', level=logging.DEBUG)
logging.warning('Hi, INTERNAL STATUS. Operating System Starting')
model = Sequential()
model.add(Dense(256, input_shape=(8,), activation='relu'))
model.add(Dense(128, activation='relu',kernel_regularizer= 'l2'))
model.add(Dense(64, activation='relu',kernel_regularizer= 'l2'))
model.add(Dense(64, activation='relu',kernel_regularizer= 'l2'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='BinaryCrossentropy', optimizer='Adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(X, y, epochs=50, batch_size=8, verbose=1)
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
print("Quintessa ACTIVE")
correct = 0
incorrect = 0

print(2*"\n")
for i in range(768):
    if predictions[i] == y[i]:
        correct += 1
    else:
        incorrect +=1

def print_data():
    if predictions[i] == y[i]:
        correct += 1
        print(Fore.GREEN +"                                                         ---{[[ Prediction CORRECT ]]}---")
        print(3*"\n")
        print('                                                                        {%s}'%i)
        print("                                    --------------------------------Case  Number------------------------------")
        print(Style.RESET_ALL)
        print("\n" + Fore.YELLOW)
        print("----------------------------------------------------------------------------------------------------------------------------------------------------------")
        print(Fore.BLUE +"            DATABASE",Fore.LIGHTMAGENTA_EX+"                                                                                                                 QUINTESSA")
        print("\n",Fore.YELLOW +"           [Test]                  --{%s}--" % y[i],
        "                                                             --{%d}--                 [Prediction]"% predictions[i])
        print("\n",Fore.BLUE +"           VALUE",Fore.LIGHTMAGENTA_EX+"                                                                                                                        VALUE",'\n')
        print(Fore.YELLOW +"----------------------------------------------------------------------------------------------------------------------------------------------------------")
        print("\n")
        logging.info('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
        print(Fore.GREEN)
        print("                                    --------------------------------  Case  --------------------------------")
        print(Style.RESET_ALL)
        print(12*"\n")
    else:
        incorrect +=1
        print(Fore.RED +"                                                         ---{[[ Prediction INCORRECT ]]}---")
        print(6*"\n")
        print('                                                                        {%s}'%i)
        print("                                    --------------------------------Case  Number------------------------------")
        print(Style.RESET_ALL)
        print("\n" + Fore.YELLOW)
        print("----------------------------------------------------------------------------------------------------------------------------------------------------------")
        print(Fore.BLUE +"            DATABASE",Fore.LIGHTMAGENTA_EX+"                                                                                                                 QUINTESSA")
        print("\n",Fore.YELLOW +"           [Test]                  --{%s}--" % y[i],
        "                                                             --{%d}--                 [Prediction]"% predictions[i])
        print("\n",Fore.BLUE +"           VALUE",Fore.LIGHTMAGENTA_EX+"                                                                                                                        VALUE",'\n')
        print(Fore.YELLOW +"----------------------------------------------------------------------------------------------------------------------------------------------------------")
        print("\n")
        logging.info('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
        print(Fore.RED)
        print("                                    --------------------------------  Case  --------------------------------")
        print(Style.RESET_ALL)
        print(12*"\n")
print("Total",Fore.LIGHTGREEN_EX+"Correct: [%d]" % correct, Style.RESET_ALL)
print("Total",Fore.LIGHTRED_EX+"Incorrect: [%d]" % incorrect)
total = 100 * correct / 768
print(Style.RESET_ALL,"  ACCURACY ",Fore.LIGHTYELLOW_EX+"%","[%d]" % total,Style.RESET_ALL)
# summarize the first 10 cases
logging.warning("Total Correct")
logging.info(correct)
logging.warning("Total Incorrect")
logging.info(incorrect)
with open("QUINTESSA.txt", 'w') as q:
    q.write("Ne")
    for i in range(768):
        if predictions[i] == y[i]:
            q.write(str(i))
            q.write(" ")
            q.write("Correct")
            q.write("\n")
            correct += 1
        else:
            q.write(str(i))
            q.write(" ")
            q.write("Incorrect")
            q.write("\n")
            incorrect +=1
    q.write("\n")
    q.write(str(correct))
    q.write(str(incorrect))
```

```python
from googlesearch import search
from bs4 import BeautifulSoup
import requests

def google_search(query):
    for url in search(query, num_results=3):
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        summary = soup.find('p').get_text()
        return summary

query = "lebron james"
print(google_search(query))
```
